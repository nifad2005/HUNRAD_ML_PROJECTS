```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


# ইনপুট ডেটা (যেমন: পড়াশোনার সময়)
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1) # .reshape(-1, 1) কারণ sklearn 2D ইনপুট চায়

# আউটপুট ডেটা (যেমন: প্রাপ্ত নম্বর)
y = np.array([2, 4, 5, 4, 5, 7])

print("ইনপুট ডেটা (X):\n", X)
print("আউটপুট ডেটা (y):\n", y)



# লিনিয়ার রিগ্রেশন মডেল তৈরি করো
model = LinearRegression()

# মডেলকে ডেটার ওপর ফিট (ট্রেন) করো
model.fit(X, y)
# নতুন ইনপুট ডেটার জন্য প্রেডিকশন করো
new_X = np.array([[7], [8]]) # ধরা যাক, ৭ ঘন্টা এবং ৮ ঘন্টা পড়াশোনার পর নম্বর
predictions = model.predict(new_X)

print(f"\nনতুন ইনপুট {new_X[0][0]} এর জন্য প্রেডিকশন: {predictions[0]:.2f}")
print(f"নতুন ইনপুট {new_X[1][0]} এর জন্য প্রেডিকশন: {predictions[1]:.2f}")

print("\nমডেল ট্রেনিং সম্পন্ন হয়েছে!")
print(f"মডেলের ঢাল (Coefficient): {model.coef_[0]:.2f}") # রেখার ঢাল
print(f"মডেলের y-ছেদক (Intercept): {model.intercept_:.2f}") # রেখা y-অক্ষকে কোথায় ছেদ করে


plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='আসল ডেটা') # আসল ডেটা পয়েন্টগুলো প্লট করো
plt.plot(X, model.predict(X), color='red', label='রিগ্রেশন রেখা') # মডেলের প্রেডিকশন রেখা প্লট করো
plt.scatter(new_X, predictions, color='green', marker='X', s=100, label='নতুন প্রেডিকশন') # নতুন প্রেডিকশন পয়েন্টগুলো প্লট করো

plt.title('লিনিয়ার রিগ্রেশন উদাহরণ')
plt.xlabel('ইনপুট (X)')
plt.ylabel('আউটপুট (y)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='আসল ডেটা') # আসল ডেটা পয়েন্টগুলো প্লট করো
plt.plot(X, model.predict(X), color='red', label='রিগ্রেশন রেখা') # মডেলের প্রেডিকশন রেখা প্লট করো
plt.scatter(new_X, predictions, color='green', marker='X', s=100, label='নতুন প্রেডিকশন') # নতুন প্রেডিকশন পয়েন্টগুলো প্লট করো

plt.title('লিনিয়ার রিগ্রেশন উদাহরণ')
plt.xlabel('ইনপুট (X)')
plt.ylabel('আউটপুট (y)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
